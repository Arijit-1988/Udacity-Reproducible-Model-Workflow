# Exploratory Data Analysis
This notebook downloads the latest sample artifact from Weights and Biases, profiles the data, applies basic cleaning, and verifies the expected schema changes.

```python
import pandas as pd
import wandb
import ydata_profiling
```

```python
run = wandb.init(project="nyc_airbnb", group="eda", save_code=True)
local_path = run.use_artifact("sample.csv:latest").file()
df = pd.read_csv(local_path)
df.head()
```

```python
profile = ydata_profiling.ProfileReport(df, title="NYC Airbnb EDA")
profile.to_widgets()
```

## Cleaning steps
1. Keep prices in the stakeholder-approved range.
2. Parse last_review as datetime.
3. Keep only points inside the NYC bounding box.

```python
min_price = 10
max_price = 350
idx = df["price"].between(min_price, max_price)
df = df[idx].copy()
df["last_review"] = pd.to_datetime(df["last_review"])
idx = df["longitude"].between(-74.25, -73.50) & df["latitude"].between(40.5, 41.2)
df = df[idx].copy()
df.info()
```
